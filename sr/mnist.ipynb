```python
import numpy as np

from util import load_mnist_data
from sr import SoftmaxRegression

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data set
train_set, valid_set, test_set = load_mnist_data("mnist.pkl.gz")
```

```python
train_X = train_set[0]
train_Y = np.zeros([train_X.shape[0], 10])
train_Y[np.arange(train_X.shape[0]), train_set[1]] = 1

test_X = test_set[0]
test_Y = np.zeros([test_X.shape[0], 10])
test_Y[np.arange(test_X.shape[0]), test_set[1]] = 1
```

```python
# let's show some pictures
show_imgs = train_X[::10000]
show_labels = train_set[1][::10000]
m = len(show_imgs)
for i in range(len(show_imgs)):
    img = show_imgs[i].reshape([28, 28])
    plt.subplot(1, m, i+1)
    plt.imshow(img, cmap='gray')
print(show_labels.reshape(-1))
```

```python
sr_clf = SoftmaxRegression(learning_rate=0.01, max_iter=10000)
sr_clf.fit(train_X, train_Y)
```

```python
test_acc = sr_clf.score(test_X, test_Y)
print("test accurcy: " + str(test_acc))
```
